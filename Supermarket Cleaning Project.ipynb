##### **1. Count Outlier**

```python
def count_outlier(df, column_name):
    # Ensure the column is numeric before processing
    if pd.api.types.is_numeric_dtype(df[column_name]):
        q3 = df[column_name].quantile(0.75)
        q1 = df[column_name].quantile(0.25)
        iqr = q3 - q1
        upper_whisker = q3 + 1.5 * iqr
        lower_whisker = q1 - 1.5 * iqr
        
        if upper_whisker == 0 or upper_whisker == 1:
            return 0
        else:
            count_outlier = int(len(df[(df[column_name] < lower_whisker) | (df[column_name] > upper_whisker)]))
            return count_outlier
    
    else:
        return 0 
```

Above is a function to count how many outliers

##### **2. Removing Outlier**

```python
def remove_outlier(df, column_name):

    q3 = df[column_name].quantile(0.75)
    q1 = df[column_name].quantile(0.25)
    iqr = q3 - q1
    upper_whisker = q3 + 1.5 * iqr
    lower_whisker = q1 - 1.5*iqr
    
    df_filtered =  df[(df[column_name] <= upper_whisker) & (df[column_name] >= lower_whisker)]
    return df_filtered
```

Above is a function to remove outlier

##### **3. Create Bar Plot**

```python
def create_bar_plot(ax, data, title, xlabel, ylabel, rotation=45):
    data.plot(kind='bar', ax=ax)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=rotation)
```
